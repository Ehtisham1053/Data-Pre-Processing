```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("audi.csv")
```

```python
df.head(5)
```

# Before applying the transformation

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,4))
plt.style.use('ggplot')
sns.distplot(df['price'])
plt.show()
```

the above grapgh shows that it is right skwed data so we will apply the log tranformation to this data


# After applying the transformation

```python

import matplotlib.pyplot as plt
import numpy as np
# applying the transformation
df['price'] = np.log1p(df['price'])

plt.figure(figsize=(8,4))
plt.style.use('ggplot')
sns.distplot(df['price'])
plt.show()
```

by applying the log1p transformation we have seen that now our transformation is somehow looking normal

# same doing for the other columns

```python

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

columns_to_plot = [ 'mileage', 'tax', 'mpg']


for i, col in enumerate(columns_to_plot):

    sns.distplot(df[col], ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')


plt.tight_layout()
plt.show()
```

```python
df['mileage'] = np.log1p(df['mileage'])
df['tax'] = np.log1p(df['tax'])
df['mpg'] = np.log1p(df['mpg'])
```

```python

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

columns_to_plot = [ 'mileage', 'tax', 'mpg']


for i, col in enumerate(columns_to_plot):

    sns.distplot(df[col], ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')


plt.tight_layout()
plt.show()
```

The above math transformation has proved that by appplying the log transfromation we have converted the right skewed data into normal data
