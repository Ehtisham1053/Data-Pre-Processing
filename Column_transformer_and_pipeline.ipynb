```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python

df = pd.read_csv("audi.csv")
```

```python
df.head(4)
```

```python
x = df.drop("price", axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head(3)
```

```python
# one hot encoding
c1 = ColumnTransformer(transformers=[
    ("one_hot", OneHotEncoder(sparse_output=False, drop="first" , handle_unknown="ignore"), [0,2,4])
],remainder='passthrough')
```

```python
# scaling of selected column
c2 = ColumnTransformer(transformers=[
    ("standard", StandardScaler(), [1,3,5, 6, 7])
],remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ("col_trans", c1),
    ("col_trans2", c2),
    ("model", LinearRegression())
])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python

from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
r2
```
