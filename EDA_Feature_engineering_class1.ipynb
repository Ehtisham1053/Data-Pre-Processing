# Encoding the catagorical columns


```python
df = pd.read_csv("audi.csv")
df.head(2)
```

```python
# prompt: one hot encode the transmission and fuelType column using the sklearn one hot encoding with drop first

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Assuming df is your DataFrame as shown in the previous code
# ... (your previous code) ...

# Create an instance of OneHotEncoder with drop='first'
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')  # sparse=False for dense output

# Fit and transform the 'transmission' and 'fuelType' columns
encoded_data = encoder.fit_transform(df[['transmission', 'fuelType']])

# Create a new DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['transmission', 'fuelType']))

# Concatenate the encoded columns with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original 'transmission' and 'fuelType' columns
df = df.drop(['transmission', 'fuelType'], axis=1)

```
